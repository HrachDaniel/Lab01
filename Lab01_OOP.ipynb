```python
import math

# Завдання 1: Обчислення математичного виразу
def task1():
    # Отримуємо введення від користувача та перетворюємо у дробове число
    x = float(input("Введіть x: "))
    # Обчислюємо за формулою: x - (x^3 / 3) + (x^5 / 5)
    result = x - (x**3 / 3) + (x**5 / 5)
    # Виводимо результат на екран
    print(f"Результат виразу: {result}")

# Виклик завдання
task1()
```

```python
import math

def task2():
    # Зчитуємо необхідні параметри прогресії
    a1 = float(input("Перший член (a1): "))
    d = float(input("Різниця (d): "))
    n = int(input("Кількість членів (n): "))
    # Використовуємо стандартну формулу суми n членів
    sn = (n * (2 * a1 + (n - 1) * d)) / 2
    # Виводимо результат на екран
    print(f"Сума прогресії: {sn}")

# Виклик завдання
task2()
```

```python
import math

def task3():
    # Вводимо довжини трьох сторін
    a = float(input("Сторона a: "))
    b = float(input("Сторона b: "))
    c = float(input("Сторона c: "))
    # Порівнюємо всі сторони між собою (повертає True, якщо всі рівні)
    is_equilateral = (a == b == c)
    # Виводимо результат на екран
    print(is_equilateral)

# Виклик завдання
task3()
```

```python
import math

def task5():
    n_limit = int(input("Введіть N: "))
    print(f"Числа Піфагора менші за {n_limit}:")
    # Перебираємо всі можливі варіанти для сторін a та b
    for a in range(1, n_limit):
        for b in range(a, n_limit):
            # Обчислюємо гіпотенузу c
            c = math.sqrt(a**2 + b**2)
            # Перевіряємо, чи є c цілим числом і чи не перевищує ліміт N
            if c < n_limit and c.is_integer():
                # Виводимо результат на екран
                print(f"{a}, {b}, {int(c)}")

# Виклик завдання
task5()
```

```python
import math

def task6():
    # Перетворюємо введений рядок у список цілих чисел
    nums = list(map(int, input("Введіть числа через пробіл: ").split()))
    if not nums: return

    # Знаходимо позиції (індекси) найменшого та найбільшого чисел
    idx_min = nums.index(min(nums))
    idx_max = nums.index(max(nums))

    # Визначаємо межі для зрізу (враховуємо, що min може бути після max)
    start = min(idx_min, idx_max)
    end = max(idx_min, idx_max)

    # Робимо зріз списку та рахуємо суму елементів у ньому
    total_sum = sum(nums[start : end + 1])
    # Виводимо результат на екран
    print(f"Сума між min та max (включно): {total_sum}")

# Виклик завдання
task6()
```

```python
import math

def task7():
    s = input("Введіть рядок з двокрапкою: ")
    # Шукаємо індекс першого входження двокрапки
    position = s.find(':')
    # Якщо символ знайдено (індекс не -1), виводимо його позицію
    if position != -1:
        print(f"Кількість символів перед ':': {position}")
    else:
        print("Двокрапка не знайдена.")

# Виклик завдання
task7()
```
